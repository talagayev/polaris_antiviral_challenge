```python
df_hlm = pd.read_csv('/mdspace/polaris_ml_challenge/data/admet/ADMET_HLM_CONFIDENTIAL.csv', delimiter=',')
df_logd = pd.read_csv('/mdspace/polaris_ml_challenge/data/admet/ADMET_LogD_CONFIDENTIAL.csv', delimiter=',')
df_KSOL = pd.read_csv('/mdspace/polaris_ml_challenge/data/admet/ADMET_KSOL_CONFIDENTIAL.csv', delimiter=',')
df_mdr1 = pd.read_csv('/mdspace/polaris_ml_challenge/data/admet/ADMET_MDR1-MDCK2_CONFIDENTIAL.csv', delimiter=',')
df_mlm = pd.read_csv('/mdspace/polaris_ml_challenge/data/admet/ADMET_MLM_CONFIDENTIAL.csv', delimiter=',')
```

```python
df_hlm = df_hlm.drop_duplicates(subset=['Molecule Name', 'CXSMILES (CDD Compatible)'])
df_KSOL = df_KSOL.drop_duplicates(subset=['Molecule Name', 'CXSMILES (CDD Compatible)'])
df_logd = df_logd.drop_duplicates(subset=['Molecule Name', 'CXSMILES (CDD Compatible)'])
df_mdr1 = df_mdr1.drop_duplicates(subset=['Molecule Name', 'CXSMILES (CDD Compatible)'])
df_mlm = df_mlm.drop_duplicates(subset=['Molecule Name', 'CXSMILES (CDD Compatible)'])
print(len(df_hlm), len(df_KSOL), len(df_logd), len(df_mdr1), len(df_mlm))
```

```python
# Start merging
df_merged = df_hlm.merge(df_KSOL, on=['Molecule Name', 'CXSMILES (CDD Compatible)'], how='outer')
df_merged = df_merged.merge(df_logd, on=['Molecule Name', 'CXSMILES (CDD Compatible)'], how='outer')
df_merged = df_merged.merge(df_mdr1, on=['Molecule Name', 'CXSMILES (CDD Compatible)'], how='outer')
df_merged = df_merged.merge(df_mlm, on=['Molecule Name', 'CXSMILES (CDD Compatible)'], how='outer')
df_merged
```

```python
# rename column in the df
df_merged = df_merged.rename(columns={'CXSMILES (CDD Compatible)':'CXSMILES'})
```

```python
moleculesProcessed = MoleculePreprocessorExtended.init_with_smiles(list(df_merged.CXSMILES))
```

```python
from rdkit import RDLogger
moleculesProcessed.csp_wash()
```

```python
preprocessedSmilesDict = moleculesProcessed.get_rawsmiles_smiles_dict()
```

```python
preprocessedSmiles_df = pd.DataFrame(list(preprocessedSmilesDict.items()), columns=['rawSmiles','preprocessedSmiles'])
```

```python
preprocessed_df = pd.merge(df_merged, preprocessedSmiles_df, 
                           left_on='CXSMILES', right_on='rawSmiles')
```

```python
preprocessed_df.dropna(subset=['preprocessedSmiles'],inplace=True) 
```

```python
preprocessed_df = preprocessed_df[preprocessed_df['preprocessedSmiles']!='']
```

```python
len(preprocessed_df)
```

```python
def remove_stereochemistry(smiles):
    # Convert the SMILES string to an RDKit molecule object
    mol = Chem.MolFromSmiles(smiles)
    
    # Use the Chem.RemoveStereochemistry function to remove stereo informationz
    Chem.RemoveStereochemistry(mol)
    
    # Convert the molecule back to a SMILES string without stereochemistry
    return Chem.MolToSmiles(mol)

preprocessed_df['preprocessedSmiles'] = preprocessed_df['preprocessedSmiles'].apply(remove_stereochemistry)
preprocessed_df
```

```python
preprocessed_df_useful = preprocessed_df[['preprocessedSmiles','in-vitro_HLM_bienta: CLint (Num) (uL/min/mg)', 'in-vitro_KSOL-PBS_bienta: mean_solubility (Num) (uM)', 'in-vitro_LogD_bienta: LogD (Num)', 'in-vitro_MDR1-MDCKII-Papp_bienta: mean_Papp_A_to_B (Num) (10^-6 cm/s)', 'in-vitro_MLM_bienta: CLint (Num) (uL/min/mg)']]
preprocessed_df_useful
```

```python
# train test split
from sklearn.model_selection import train_test_split
train, test = train_test_split(preprocessed_df_useful, test_size= 0.2, random_state=42)
test = test[~test['preprocessedSmiles'].isin(train['preprocessedSmiles'])]
print(len(train), len(test))
```

```python
train.to_csv('./train_test_split/train_admet.csv', index=False, header=True)
test.to_csv('./train_test_split/test_admet.csv', index=False, header=True)
preprocessed_df_useful.to_csv('./train_test_split/full_admet.csv', index=False, header=True)
```
